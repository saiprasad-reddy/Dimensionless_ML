<h3>1. Draw a scatter plot showing the relationship between the horsepower and price of the
automobiles.</h3>

```python
plt.figure()
plt.scatter(Auto.horsepower,Auto.price,s=50,c='red')
```

<h3> 2. Write a code to generate following graph on automobile dataset using pandas and
matplotlib.</h3>

```python
plt.figure()
plt.hist(Auto['fuel-type']);
plt.xlabel('Fuel Type')
plt.ylabel('Number of Automobiles')
plt.xticks()
plt.show()
```

<h3>3. Draw a histogram on normalized lossess of automobiles</h3>

<h3>4. Write a code to generate following graph on automobile dataset using pandas, matplotlib
and seaborn.</h3>

```python
plt.figure()
x1 = list(Auto[Auto['body-style']=='convertible']['make'])
x2 = list(Auto[Auto['body-style']=='hatchback']['make'])
x3 = list(Auto[Auto['body-style']=='sedan']['make'])
x4 = list(Auto[Auto['body-style']=='wagon']['make'])
x5 = list(Auto[Auto['body-style']=='hardtop']['make'])

labels=['convertible','hatchback','sedan','wagon','hardtop']
plt.hist([x1, x2, x3, x4, x5],
         label=labels
         );
# plt.xticks(Lbins) # to set the ticks according to the bins
plt.xlabel('Make')
plt.xticks(rotation=90)
plt.ylabel('Number of Automobiles')
plt.legend(loc=2);
plt.show()
```

<h3>5. Draw a boxplot for price of automobiles based on body-style using seaborn.</h3>

```python
sns.boxplot(x='body-style',y='price',data=Auto,hue='body-style');
```
