<h2>Seaborn</h2>
<h4>Seaborn is a python data visualization library based on matplotlib. It provides high-level interface for drawing attartive and informative stastical graphics</h4>

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.__version__
```

<h3>Line Plots</h3>

```python
sns.lineplot(x="petal_length",y="petal_width",data=iris,hue='species')
```

```python
graph=sns.lineplot(x="petal_length", y="petal_width",data=iris)
graph.set_title("Line Plot for Petal Length")
graph.set(xlabel = "Petal Length", ylabel = "Petal Width")
graph.set_xlim(0,10)
graph.set_ylim(0,5)
```

<h3>Scatter Plots</h3>

```python
sns.scatterplot(x="petal_length",y="petal_width",data=iris,hue='species'
)
#hue used for categorical data to sperate the different species.
```

<h3>Histograms</h3>

```python
sns.distplot(iris.sepal_length)
```

<h3>Kernal density Estimation (kde plots)</h3>

```python
sns.distplot(iris.sepal_length,hist=False)
```

```python
sns.distplot(iris.sepal_length,kde=False)
```

<h3>Bar Chatrs</h3>

```python
sns.barplot(x='species',y='petal_width',data=iris)
#black line is error interval, generally denoted in barcharts
```

```python
sns.barplot(x='petal_width',y='species',data=iris,orient='h');
```

<h3>Box plots</h3>

```python
sns.boxplot(data=iris);
```

```python
sns.boxplot(x='species',y='sepal_width',data=iris,hue='species');
```
